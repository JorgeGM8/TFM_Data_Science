```python
import pandas as pd
from skrub import TableReport
```

Lectura de csv

```python
df_autoinvestor = pd.read_csv('../data/processed/autoinvestor_procesado.csv')
df_demograficos = pd.read_csv('../data/processed/datos_demograficos_ine_procesado.csv')
```

Unión de dataframes y reporte inicial

```python
df_merge = df_autoinvestor.merge(
    df_demograficos,
    left_on=['LOCATIONNAME', 'YEAR'],
    right_on=['Distritos', 'Periodo'],
    how='left'
)
print('\n--> Unidos dataframes de autoinvestor y datos demográficos\n')
```

```python
TableReport(df_merge)
```

```python
df_merge = df_merge.drop(['Distritos', 'Periodo', 'Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)
print('\n--> Eliminadas columnas extra\n')
```

Guardar reporte final

```python
report = TableReport(df_merge)

# Guardar reporte HTML
with open('../reports/report_merge.html', 'w') as f:
    f.write(report.html())
print('\n--> Reporte guardado en reports/report_merge.html\n')
```

Guardar csv procesado

```python
df_merge.to_csv('../data/processed/merge_autoinvestor_demograficos.csv')
print('\n--> Datos procesados guardados en data/processed/merge_autoinvestor_demograficos.csv\n')
```
